## Gerador de QR CODE de contato comercial

```python
import qrcode

def gerar_qr_contato_comercial(nome, cargo, empresa, email, telefone):
    # Informações de contato formatadas
    contato = f"BEGIN:VCARD\nVERSION:3.0\nFN:{nome}\nORG:{empresa}\nTITLE:{cargo}\nEMAIL:{email}\nTEL:{telefone}\nEND:VCARD"

    # Criar o objeto QRCode
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )

    # Adicionar os dados ao QRCode
    qr.add_data(contato)
    qr.make(fit=True)

    # Criar uma imagem QRCode usando a biblioteca PIL
    img = qr.make_image(fill_color="black", back_color="white")

    # Salvar a imagem ou exibir
    img.save("contato_comercial.png")
    img.show()

# Exemplo de uso
nome = input("Escreva seu NOME:")
cargo = input("Escreva seu CARGO:")
empresa = input("Escreva seu Empresa:")
email = input("Escreva seu E-MAIL:")
telefone = input("Escreva seu Telefone:")

gerar_qr_contato_comercial(
    nome= nome,
    cargo= cargo,
    empresa= empresa,
    email= email,
    telefone= telefone,
)
```
